```python
def randomiseMissingData(df2):
    "randomise missing data for DataFrame (within a column)"
    df = df2.copy()
    for col in df.columns:
        data = df[col]
        mask = data.isnull()
        samples = random.choices( data[~mask].values , k = mask.sum() )
        data[mask] = samples

    return df


import random
import numpy as np
merge = pd.concat([df1[['M1','P1']], df2[['M2','P2']], df3[['M3','P3']]],axis=1)
merge = randomiseMissingData(merge)
merge.head()
```

```python
#first: equalize intensity of each column to mean sum intesity of the lfq channels for all experiments
exp1_raw = data.iloc[:,0:2]
exp2_raw = data.iloc[:,2:4]
exp3_raw = data.iloc[:,4:6]

target = np.mean(pd.concat([exp2_raw.sum(axis=0), exp3_raw.sum(axis=0)]))
print('mean sum intensity', target)
```

```python
#normalization factors for first experiment
norm_facs = target / exp1_raw.sum(axis=0)
exp1_sl = exp1_raw.multiply(norm_facs, axis=1)
```

```python
#normalization factors for second experiment
norm_facs = target / exp2_raw.sum(axis=0)
exp2_sl = exp2_raw.multiply(norm_facs, axis=1)
```

```python
#normalization factors for third experiment
norm_facs = target / exp3_raw.sum(axis=0)
exp3_sl = exp3_raw.multiply(norm_facs, axis=1)
```

```python
#normalized experiment
data_sl =pd.concat([ exp2_sl, exp3_sl], axis=1)

fig,ax=plt.subplots(figsize=(12,6))
g = sns.boxplot(data=np.log1p(data_sl),ax=ax,palette=my_pal[:4])
g.set_xticklabels(data_sl.columns, rotation=30, ha='right')
plt.title('norm data')
plt.xlabel('samples')
plt.ylabel('log2 of intensity')
plt.show()
```

```python
%%R
row.names(limma_data) <- accession # add accessions as row names
# set up the design matrix
group <- as.factor(c(rep("senescent", 3), rep("control", 3)))
group <- factor(group, levels(group)[c(2, 1)]) # set the factor order
design <- model.matrix(~ 0 + group)
colnames(design) <- c("senescent", "control")

# make the contrast
contrast <- makeContrasts(senescent-control, levels = design)

S <- 1:3
C <- 4:6
# do the linear model fitting on log2 values
limma_log_data <- log2(limma_data[c(S, C)])
fit <- lmFit(limma_log_data, design)

# get the fit for the contrast of interest
fit2 <- contrasts.fit(fit, contrast)

# do the empirical Bayes moderation of the test statistic (with trended variance)
fit2 <- eBayes(fit2, trend = TRUE)

# grab the information in topTable so we can get the data to plot candidates
# the coef parameter has to do with the contrast of interest
# specify no sorting of results and a number that is longer than the data table
tt_limma <- topTable(fit2, coef = 1, sort.by = "none", number = Inf)
```

```python
# let's see how many up and down candidates, and the top tags
%R topTable(fit2)
```

```python
%%R
# statisticl test results
limma_data$Acc <- accession
limma_data$PValue <- tt_limma$P.Value
limma_data$FDR <- tt_limma$adj.P.Val
head(limma_data)
# add a DE candidate status column
limma_data$candidate <- cut(limma_data$FDR, breaks = c(-Inf, 0.01, 0.05, 0.10, 1.0), 
                           labels = c("high", "med", "low", "no"))
```

```python
irs = pd.concat([ exp2_sl.sum(axis=1), exp3_sl.sum(axis=1)], axis=1)
irs.columns = ["sum2", "sum3"]
irs.head()
```

```python
#mean of the sum intensity of all the proteins in each channel
irs = pd.concat([exp2_sl.sum(axis=1), exp3_sl.sum(axis=1)], axis=1)
irs.columns = ["sum2", "sum3"]

#geometric mean of the sum intensity of all the proteins
irs['average']=np.exp(np.log(irs).mean(axis=1))
irs['fac2']=irs['average'].values / irs['sum2'].values
irs['fac3']=irs['average'].values / irs['sum3'].values
irs.head()
```

```python
#apply normalization factors for eaqch protein in dataset 1 and 2

data_irs_2 = exp2_sl.multiply(irs.fac2, axis=0)
data_irs_3 = exp3_sl.multiply(irs.fac3, axis=0)
#combine again
data_irs = pd.concat([data_irs_2,data_irs_3],axis=1)
data_irs.head()
```
